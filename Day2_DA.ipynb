```python
#after vertically stacking
A = np.array([32,35,78,89,34,56,45,67,23]).reshape(3,3)
B = np.array([45,78,45,56,34,45,56,78,24]).reshape(3,3)
print(A)
print("_________________")
print(B)
print("After vertically stacking the arrays!")
print(np.vstack((A,B)))
```

```python
#after dstack
A = np.array([32,35,78,89,34,56,45,67,23]).reshape(3,3)
B = np.array([23, 45, 67, 78, 56,67,45,89,45]).reshape(3,3)
print(A)
print(B)
print("After dstack")
print(np.dstack(A))
print(np.dstack(B)) #no of rows -->no of groups
#no of col in every group --> no of rows in every group
```
