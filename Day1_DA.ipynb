```python
#get the salary from the user if the salary is more than or equal to 50000 applay 10% tax reduction
#if the salary is less than 50000 apply 5% tax
#if the salary is more than 70000 aplly 12 % tax
#if the salary exceeds 100000 you are promted to next level
n=float(input("enter salary"))
if n>=50000 and n<70000:
    i=n*(10/100)
    print("salary",n-i)
elif n<50000:
    i=n*(5/100)
    print("salary",n-i)
elif n>=70000 and n<100000:
    i=n*(12/100)
    print("salary",n-i)
else:
    print("you are promoted to next level")
```
