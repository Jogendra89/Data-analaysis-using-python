```python
#write a plotting statements there are 5 subjects psychology,nanotechonlogy,medicine,anthapology,geology
#two students have applied to do masters in any of these 5 fields with lineplot show in which field they can excel the cutoff for each subject is 70
Sub=["psy","nano","med","anth","geo"] #subjects
S1=[67,45,73,71,57] #student1
S2=[43,67,68,53,85] #student2
cutoff=[70,70,70,70,70] #cutoff
plt.plot(Sub,cutoff,label="Cutoff",color="k",marker="P",linestyle=":",linewidth=3)
plt.plot(Sub,S1,label="S1",color="b",marker="H")
plt.plot(Sub,S2,label="S2",color="r")
plt.title("Marks of Masters students")
plt.legend()
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
#ind vs aus match is
#ind has given 298 as target for australia
#runs per every 5overs is given as follows
#25,51,84,131,160,189,220,250,267,298
#aus had completed only 25 overs
#runs per every 5overs ids given as follows
#15,41,94,110,151
#plot overs uvs runs for the both countries with in as blue line and aus yellow line,give appropriate labels and legend ,the marker for ind is flagand for aus is plus
#take overs from arange.Create runs as array
import numpy as np
Oversind=np.arange(5,51,5)
Oversaus=np.arange(5,26,5)
Ind=np.array([25,51,84,131,160,189,220,250,267,298])
Aus=np.array([15,41,94,110,151])
plt.plot(Oversind,Ind,label="Ind",color="b",marker=">",linewidth="2")
plt.plot(Oversaus,Aus,label="Aus",color="y",marker="+",linewidth="2")
plt.title("Score of ind vs aus match")
plt.legend()
plt.grid()
plt.xlabel("Overs")
plt.ylabel("Runs")
```
