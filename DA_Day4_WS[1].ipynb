```python
from bs4 import BeautifulSoup
import requests
s="https://en.m.wikipedia.org/wiki/List_of_tallest_buildings_in_India"
page=requests.get(s)
code=BeautifulSoup(page.text,'html')
h=code.find('table')
p=h.find_all('th')
headings=[i.text for i in p]
print(headings)

```

```python
import pandas as pd
df=pd.DataFrame(columns=headings)
print(df)
rows=h.find_all('td')
l=[i.text for i in rows]
print(l)
```

```python
st=0
sp=9
for i in range(len(l)//9):
    df.loc[i]=l[st:sp]
    st+=9
    sp+=9
df
```

```python
df.to_csv("buildings.csv")
```
